```python
X = data.year.values
X = X.reshape(47,1)
X.shape
```

```python
Y = data.pci.values
Y
```

```python
# Simple Linear Regression
import sklearn.linear_model as lm
lin_reg = lm.LinearRegression()
```

```python
lin_reg.fit(X,Y)
```

```python
data['predicted_values'] = lin_reg.predict(X)
```

```python
data[['year','pci','predicted_values']].head(20)
```

```python
pred_values = lin_reg.predict(X)
pred_values
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(Y,pred_values)
print('Accuracy of Model :',accuracy)
```

```python
# plotting actual values
plt.plot(X,Y,color='cyan',label='Actual Values')
plt.scatter(X,Y,color='cyan')
# plotting predicted values
plt.plot(X,pred_values,color='k',label='Predicted Values')
plt.scatter(X,pred_values,color='k')
plt.title('Pci')
plt.xlabel('Years')
plt.ylabel('Pci')
plt.legend()
plt.show()
```

```python
# Polynomial Regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg_2 = PolynomialFeatures(degree=2)
x_poly_2 = poly_reg_2.fit_transform(X)
x_poly_2
```

```python
lin_reg.fit(x_poly_2,Y)
```

```python
pred_values = lin_reg.predict(x_poly_2)
```

```python
# Plot Actual Values using polynomial regression with degree 2
plt.scatter(X,Y,color='red')
plt.plot(X,pred_values,color='green',label='Regression Line')
plt.title('Model Using Polynomial Regression Degree = 2')
plt.xlabel('Years')
plt.ylabel('PCIs')
plt.legend()
plt.show()
```

```python
# Degree = 3
poly_reg_3 = PolynomialFeatures(degree=3)
x_poly_3 = poly_reg_3.fit_transform(X)
lin_reg.fit(x_poly_3,Y)
```

```python
pred_values = lin_reg.predict(x_poly_3)
```

```python
# Plot Actual Values Using Degree 3
plt.scatter(X,Y,color='red')
plt.plot(X,pred_values,color='blue',label='Regression Line')
plt.title('Model Using Polynomial Regression Degree = 3')
plt.xlabel('Years')
plt.ylabel('PCIs')
plt.legend()
plt.show()
```

```python
# Predicting The Final Output Using Polynomial Regression Degree = 3
Year = float(input('Enter The Year : '))
predictions = lin_reg.predict(poly_reg_3.fit_transform([[Year]]))
print('For Year',Year,'PCI Will Be',int(predictions[0]))
```
