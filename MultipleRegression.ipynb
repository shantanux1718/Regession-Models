```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/aman9801/multiple-linear-regression-on-housing-dataset/main/mlr_housing.csv')
data.head(10)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['mainroad'] = data['mainroad'].replace({'yes': 1, 'no': 0})
data['guestroom'] = data['guestroom'].replace({'yes': 1, 'no': 0})
data['basement'] = data['basement'].replace({'yes': 1, 'no': 0})
data['hotwaterheating'] = data['hotwaterheating'].replace({'yes': 1, 'no': 0})
data['airconditioning'] = data['airconditioning'].replace({'yes': 1, 'no': 0})
data['prefarea'] = data['prefarea'].replace({'yes': 1, 'no': 0})
furnishingstatus=LabelEncoder()
data['furnishingstatus']=furnishingstatus.fit_transform(data['furnishingstatus'])
```

```python
data.head(10)
```

```python
# Extract independent and dependent variable
x = ['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
       'parking', 'prefarea', 'furnishingstatus']
x = data[x].values
x
```

```python
y = data.price.values
y
```

```python
# Split The Dataset Into Training And Testing Subset
from sklearn.model_selection import train_test_split
# test_size=0.2 means 20% of the data will be used for Testing and remaining 80% used to training
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train, y_train)
```

```python
pred_values = mlr.predict(x_test)
pred_values
```

```python
from sklearn.metrics import r2_score
acc = r2_score(y_test, pred_values)
print("Accuracy of model is: ", acc*100)
```

```python
# Final Predictions
columns = ['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad',
           'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
           'parking', 'prefarea', 'furnishingstatus']
user_input = {column: float(input(f"Enter the value for {column}: ")) for column in columns}
user_data = pd.DataFrame([user_input])

user_pred = mlr.predict(user_data)
print("Predicted price:", user_pred[0])
```
