```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mat
import sklearn.linear_model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
bluff = pd.read_csv("https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv")
bluff
```

```python
bluff.info()
```

#simple linear regression

```python
x = bluff.Level.values[:,np.newaxis]
y = bluff.Salary.values
```

```python
lin_reg.fit(x,y)
```

```python
prevalue = lin_reg.predict(x)
mat.scatter(x,y,color='red')
mat.plot(x,prevalue,color='green',label='Regression line')
mat.title('Bluff model using SLR')
mat.xlabel('Level')
mat.ylabel('Salaries')
mat.legend()
mat.show()
```

```python
#Polynomial Regression :
#1st we increase degree of polynomial variable then we train model
```

#Polynomial Regression with degree=2

```python
#Build Model using Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
preg = PolynomialFeatures(degree=2)
xpoly2 = preg.fit_transform(x)
xpoly2

'''
when degree=2
matrix will be
col1  = 1,2,3,....
col2 = x*1
col2 = x*x
'''

'''
when degree=3
matrix will be
col1  = 1,2,3,....
col2 = x*1
col2 = x*x
col3 = x*x*x
'''

'''
when degree=4
matrix will be
col1  = 1,2,3,4....
col2 = x*1
col2 = x*x
col3 = x*x*x
col4 = x*x*x*x
'''
```

```python
lin_reg.fit(xpoly2,y)
```

```python
prevalue = lin_reg.predict(xpoly2)
mat.scatter(x,y,color='red')
mat.plot(x,prevalue,color='green',label='Regression line')
mat.title('Bluff model using Polynomial Regression with degree=2')
mat.xlabel('Level')
mat.ylabel('Salaries')
mat.legend()
mat.show()
```

#Polynomial Regression with degree=3
